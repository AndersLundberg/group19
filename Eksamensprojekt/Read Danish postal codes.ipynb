```python
import numpy as np, pandas as pd
```

```python
postalcodes = pd.read_csv('dk_postnr.csv', sep=';', header=0, 
                          names=['lau2_code', 'nation_code', 'lau2_name', 'lau2_name_latin', 'nuts3', 'nuts3_latin'])

postalcodes.head(2)

post_dict = pd.Series.to_dict(pd.Series(list(postalcodes['nuts3']), index=list(postalcodes.lau2_name)))

sorted(post_dict.items())
```

```python
ny1 = pd.read_pickle('both1.pkl')
ny2 = pd.read_pickle('both2.pkl')

df_both = pd.concat([ny1, ny2], axis=0, join='inner', ignore_index=True)

df_both.describe()
```

```python
df_both['nuts3'] = df_both['city'].map(post_dict)
df_both.describe()
```
